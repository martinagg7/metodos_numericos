```python
import numpy as np
```

```python
def euler(f, t0, tf, y0, n):
    # Variables
    h = (tf - t0) / n  # Tamaño del paso
    y = y0
    t = t0

    # Creamos dos listas para almacenar los valores
    t_values = [t0]
    y_values = [y0]

    # Bucle para realizar los pasos del método de Euler
    while t < tf:
        # Actualización de y utilizando la fórmula correcta del método de Euler
        y = y + h * f(t, y)

        # Actualización del valor de t
        t = t + h

        # Guardamos los valores en las listas
        t_values.append(t)
        y_values.append(y)

    # Devolvemos las listas de valores después del bucle
    return t_values, y_values
```

```python
def R_4(f, t0, tf, y0, n):
    h = (tf - t0) / n
    y_values = [y0]
    t_values = [t0]
    y = y0
    t = t0

    while t < tf:
        k_1 = f(t, y)
        k_2 = f(t + h/2, y + (h/2) * k_1)
        k_3 = f(t + h/2, y + (h/2) * k_2)
        k_4 = f(t + h, y + h * k_3)
        y = y + (h/6) * (k_1 + 2 * k_2 + 2 * k_3 + k_4)

        # Actualización del valor de t
        t = t + h

        # Guardamos los valores
        y_values.append(y)
        t_values.append(t)

    return t_values, y_values
```

```python
def heun(f,t0,tf,y0,n):
  h=(tf-t0)/n
  y=y0
  t=t0

  #nos creamos dos listas para guardar los valores y luego poder representarlo
  t_values=[t0]
  y_values=[y0]

  while t<tf:
    y_euler=y+h*f(t,y)
    y=y+h/2*(f(t,y)+f(t+h,y_euler))

    #actualizamos valor t
    t=t+h

    #guardamos lista
    t_values.append(t)
    y_values.append(y)

    return t_values,y_values

```
